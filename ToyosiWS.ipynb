```python
from bs4 import BeautifulSoup as bs 
import requests 
import re
import numpy as np 
```

```python
Mechanics_info = []
for i in range(1,4):
    url = f'https://mymoto.com.ng/find-mechanic?p={i}&category=0&zoom=15&is_mile=0&directory_radius=0&view=list&sort=newest'
    site = requests.get(url).text
    soup = bs(site, 'lxml')
    page = soup.find('div', class_='sabai-directory-listings-container')

    mechos = page.find_all('div', class_='sabai-col-xs-9 sabai-directory-main')

    for mecho in mechos:
        mechos_name =  mecho.find('div', class_='sabai-directory-title').text.strip()
        #The Services offered
        if (mecho.find('div', class_='sabai-directory-category')) is not None:
            mechos_service = mecho.find('div', class_='sabai-directory-category').text.strip()
            mechos_services = re.sub(r'\s{2,3}', ',', mechos_service)
        else:
            mechos_services = np.nan

        #The Location   
        mechos_location = mecho.find('div', class_='sabai-directory-location').text.strip()
        
        #The Telephone contact
        if (mecho.find('div', class_='sabai-directory-contact-tel')) is not None:
            mechos_contact_telephone = mecho.find('div', class_='sabai-directory-contact-tel').a.text
        else:
            mechos_contact_telephone = np.nan

        #The Mobile contact 
        if (mecho.find('span', {'class':'sabai-hidden-xs'})) is not None:   
            mechos_contact_mobile = mecho.find('span', {'class':'sabai-hidden-xs'}).text
        else:
            mechos_contact_mobile = np.nan

            
        #The Email
        if (mecho.find('div', {'class':'sabai-directory-contact-email'})) is not None:
            mechos_email = mecho.find('div', {'class':'sabai-directory-contact-email'}).text.strip()
        else:
            mechos_email = np.nan
        
        #The Website
        if (mecho.find('div', {'class':'sabai-directory-contact-website'})) is not None:
            mechos_website = mecho.find('div', {'class':'sabai-directory-contact-website'}).text
        else:
            mechos_website = np.nan

        alls = mechos_name, mechos_services, mechos_location, mechos_contact_telephone,    mechos_contact_mobile, mechos_email, mechos_website
        Mechanics_info.append(alls)
        print(Mechanics_info)
```

```python
import pandas as pd
names = ['mechos_name', 'mechos_services', 'mechos_location', 'mechos_contact_telephone', 'mechos_contact_mobile', 'mechos_email', 'mechos_website']
data = pd.DataFrame(Mechanics_info)
data.columns = names
data
```

```python
data.to_excel('Mechanics.xlsx')
```
