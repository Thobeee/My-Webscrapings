```python
col = table.find_all('th')
column_names = [c.string for c in col][5:]
column_names
```

```python
table_rows = table.find_all('tr')
l = []
for tr in table_rows:
    td = tr.find_all('td')
    row = [str(tr.text).strip('\n') for tr in td]
    l.append(row)
    
#print(l[3])
df = pd.DataFrame(l, columns=["Country", "RateY", "YearY", "RateA", "YearA", "RateE", "YearE", "RateYGPI", "YearYGPI"])
```
