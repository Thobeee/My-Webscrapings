```python
import bs4
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
```

```python
my_url = 'https://en.wikipedia.org/wiki/List_of_countries_by_literacy_rate'
uClient = uReq(my_url)
page_html = uClient.read()
uClient.close()
```

```python
page_soup = soup(page_html, 'html.parser')
page_soup.h1
```

```python
page_soup.body.table
```

```python
container = page_soup.findAll('div', {'class':'mw-parser-output'})
len(container)
```

```python
contain = container[0]
contain
```
